```python
np.random.seed(12345678)
x = np.random.random(10) # 0 ~ 1 之間產生十個數
y = np.random.random(10)
print(x)
print(y)
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
print(stats.linregress(x, y)) # 顯示參數
print(slope, intercept, r, p, std_err) # 參數是傳回值，所以也可以用傳回值來顯示
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(x, y, 'o')
ax.plot(x, intercept + x * slope)
plt.show()
```
