# Question 1: Probability

Below is the family tree for a small parade of elephants, one of which (“C”) exhibits albinism caused by a mutation in the gene for skin pigmentation. Each elephant carries two copies of this gene and in order to be an albino, both copies must have the mutation. When the elephants mate, each parent passes one of their copies onto the child with equal probability (regardless of mutation).

Elephants A and B each carry only 1 copy of the mutation (and thus don’t exhibit albinism) and elephant C carries 2 copies (and is affected), as shown in the figure. Elephants C, D, and E are all descendants of A & B, and Elephant F is a descendant of D & E.

### a) Given that elephant D is not an albino, what is the probability it has exactly one copy of the mutation? [1 point]

假設正常基因是 $X$ 白化症基因是 $y$，照題目的說法 A 和 B 都是 $Xy$ 而 C 是 $yy$

因為 A 和 B 會各傳一個基因給 D 所以 D 可能的基因組合是 $XX$, $Xy$, $yX$, $yy$，但是 D 沒有白化症所以只剩下三種可能 $XX$, $Xy$, $yX$，而這三種內只有 $Xy$ 和 $yX$ 是帶有剛好一個白化症基因的，因此機率是 $P=\frac{2}{3}$

### b) Given that elephant E is not an albino, what is the probability it has exactly one copy of the mutation? [1 point]

答案根原理同上 $P=\frac{2}{3}$

### c) What is the probability that elephant F, when born, will be an albino? [2 points]

D 和 E 可能是 $XX$, $Xy$, $yX$ 所以有 $3 \times 3 = 9$ 種組合，但是每一種組合內又可以產生 $2\times 2 = 4$ 種可能，因此 F 總共有 $9\times 4 = 36$ 種可能性

如果 F 要有白化症，那 (D, E) 的組合只有可能是 $(Xy, Xy)$, $(Xy, yX)$, $(yX, Xy)$, $(yX, yX)$ 四組，而且這四組內的每一種都是把 $y$ 傳給 F，所以每一組只有一種 $yy$ 的可能: $4 \times 1 = 4$

因此 F 要有白化症的可能性是 $P=\frac{4}{36} = \frac{1}{9}$.

也可以畫表格來數看看

|(D, E)                        |F (1)         |F (2)         |F (3)         |F (4)         |
|-----------------------------:|-------------:|-------------:|-------------:|-------------:|
|$(X_{D1}X_{D2}, X_{E1}X_{E2})$|$X_{D1}X_{E1}$|$X_{D1}X_{E2}$|$X_{D2}X_{D1}$|$X_{D2}X_{E2}$|
|$(X_{D1}X_{D2}, Xy)$          |$X_{D1}X$     |$X_{D1}y$     |$X_{D2}X$     |$X_{D2}y$     |
|$(X_{D1}X_{D2}, yX)$          |$X_{D1}y$     |$X_{D1}X$     |$X_{D2}y$     |$X_{D2}X$     |
|$(Xy, X_{E1}X_{E2})$          |$X_{D}X_{E1}$ |$X_{D}X_{E2}$ |$yX_{E1}$     |$yX_{E2}$     |
|$(Xy, Xy)$                    |$X_{D}X_{E}$  |$Xy$          |$yX$          |$y_{D}y_{E}$  |
|$(Xy, yX)$                    |$X_{D}y$      |$X_{D}X_{E}$  |$y_{D}y_{E}$  |$yX$          |
|$(yX, X_{E1}X_{E2})$          |$yX_{E1}$     |$yX_{E2}$     |$X_{D}X_{E1}$ |$X_{D}X_{E2}$ |

|$(yX, yX)$                    |$y_{D}y_{E}$  |$yX$          |$Xy$          |$X_{D}X_{E}$  |


### d) What is the probability that elephant F, when born, will carry exactly one copy of the mutation? [2 points]

#### 要注意 F 尚未出生

如果是數表格，$Xy$ 或 $yX$ 共有 16 組，所以機率是 $P=\frac{16}{36}=\frac{4}{9}$

如果是要用計算的，分母一樣是 36 分子就要討論
* (D, E) = $(XX, Xy)$ 或是 $(XX, yX)$ 有兩組，從 D 中選出一個 $X$ 有兩種可能，此時 E 只能選 $y$ 所以只有一種選法 --> $2 \times 2 \times 1 = 4$
* (D, E) = $(Xy, XX)$, $(Xy, Xy)$, $(Xy, yX)$
  * $(Xy, XX)$: D 只能選 y 所以是一種，E 有兩種選法 --> $1 \times 2 = 2$
  * $(Xy, Xy)$, $(Xy, yX)$: 有兩組，當 D 選定基因時 (兩種選法)，E 就只能選另一個基因 --> $2 \times 2 \times 1 = 4$
  * 所以全部的選法是 $2 + 4 = 6$
* (D, E) = $(yX, XX)$, $(yX, Xy)$, $(yX, yX)$ 原理跟上一個一樣，所以有 6 種選法
因此分子是 $4 + 6 + 6 = 16$ --> 機率 $P = \frac{16}{36} = \frac{4}{9}$

### e) Suppose that elephant F is born and is not an albino. Given this, what is the probability that elephant F carries exactly one copy of the mutation? [2 points]

#### 注意 F 已經出生

F 已經出生且沒有白化症的情況下，帶有的基因就只有$XX$, $Xy$, $yX$ 三種可能，所以有剛好一個白化症基因的機率是 $P = \frac{2}{3}$

### f) Suppose that elephant F is born and is not an albino. Given this, what is the probability that elephant D carries exactly one copy of the mutation? [2 points]

\begin{equation}
P(D 只有一個 y |F 已經出生且沒有白化症) = \frac{P(F 已經出生且沒有白化症|D 只有一個 y)P(D 只有一個 y)}{P(F 已經出生且沒有白化症)}
\end{equation}

* $P(D 只有一個 y) = \frac{2}{3}$ 注意 D 是沒有白化症的
* $P(F 已經出生且沒有白化症) = \frac{3}{4}$
* $P(F 已經出生且沒有白化症|D 只有一個 y) = \frac{20}{24} = \frac{5}{6}$
  * (D, E) = $(Xy, XX)$, $(Xy, Xy)$, $(Xy, yX)$, $(yX, XX)$, $(yX, Xy)$, $(yX, yX)$
  * 分母 = $6 \times 4 = 24$
  * 分子 = $[2\times 2 + (1_{X}\times 2_{Xy} + 1_{y}\times 1_{X}) + (1_{X} \times 2_{yX} + 1_{y} \times 1_{X})]\times 2 = (4 + 3 + 3) \times 2 = 20$
  
所以 $P(D 只有一個 y |F 已經出生且沒有白化症) = \frac{(5/6) * (2/3)}{3/4}= \frac{20}{27} $

### a) Consider 4 users, Adam, Bob, Chris and David. Adam and Chris share identical characteristics except for their equity values. Bob and David also share identical characteristics (with each other, not necessarily Adam and Chris), except for their equity values.

|Name |Equity value|Modeled Probability|
|----:|-----------:|------------------:|
|Adam | 50,000     |50%                |
|Bob  | 200,000    |50%                |
|Chris| 40,000     |?                  |
|David| 190,000    |?                  |

Based on the coefficients above, who would you think has a higher probability of buying a stock of Company X?
* Chris
* David
* They have the same probability
* Cannot tell based on the information provided

What is your reasoning? (you need not calculate an exact probability to answer this question. Just explain your reasoning in general terms.) [2 points]

\begin{equation}
\ln\frac{p}{1-p} = \sum_{i=0}^{n}\theta_{i} x_{i} = \theta_{0} x_{0} + \theta_{1} x_{1} + \theta_{2} x_{2} + \cdots + \theta_{n} x_{n}
\end{equation}

當除了 equity values 以外的全部的參數都相等時，可以比對兩個使用者在 equity value 的差距
* Chris - Adam = $\ln 40000 - \ln 50000 = \ln \frac{4}{5} = \ln 0.8 \sim -0.22$
* David - Bob = $\ln 190000 - \ln 200000 = \ln \frac{19}{20} = \ln 0.95 \sim -0.05$

因為 $\ln$ 是遞增函數，所以這一項越大的話，表示機率越大，因此 David 有比較高的機率買 X 公司的股票

### a) List the top 5 cities which had the highest number of users cancel orders on 1st Dec 2016. [2 points]

先建立兩個表格來測試結果

```SQL
SELECT u.City, COUNT(u.User_id) 
FROM users AS u 
JOIN trades AS t 
ON u.User_id = t.User_id 
WHERE DATE(t.Timestamp) = '2016-12-01' AND 
      t.Status = 'cancelled' 
GROUP BY u.City 
ORDER BY 2 DESC 
LIMIT 5;
```

````python
import random
import datetime
import pandas as pd

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = random.randrange(int_delta)
    return start + datetime.timedelta(seconds=random_second)

d1 = datetime.datetime.strptime('1/1/2010 12:00 AM', '%m/%d/%Y %I:%M %p')
d2 = datetime.datetime.strptime('1/1/2019 12:00 AM', '%m/%d/%Y %I:%M %p')

d3 = datetime.datetime.strptime('12/1/2016 12:00 AM', '%m/%d/%Y %I:%M %p')
d4 = datetime.datetime.strptime('12/7/2016 12:00 AM', '%m/%d/%Y %I:%M %p')

dict1 = {}
dict1['User_id'] = [i for i in range(1, 11)]
dict1['City'] = [random.choice(['SF', 'NY', 'LA', 'OKC', 'SLC', 'SD']) for i in range(1, 11)]
dict1['Email'] = ['email_' + str(i) for i in range(1, 11)]
dict1['Address'] = ['Address_' + str(i) for i in range(1, 11)]
dict1['Created_at'] = [random_date(d1, d2) for i in range(1, 11)]

df_users = pd.DataFrame(dict1)
# df_users

dict2 = {}
dict2['Order_id'] = [random.randint(1000,2000) for i in range(1000)]
dict2['User_id'] = [random.randint(1, 11) for i in range(1000)]
dict2['Timestamp'] = [random_date(d3, d4) for i in range(1000)]
dict2['Symbol'] = [chr(random.randint(65, 90)) for i in range(1000)]
dict2['Price'] = [random.randint(1, 10) * 1.0 for i in range(1000)]
dict2['Quantity'] = [random.randint(1, 3) for i in range(1000)]
dict2['Side'] = ['buy' if random.random() > 0.5 else 'sell' for i in range(1000)]
dict2['Status'] = ['completed' if random.random() > 0.5 else 'cancelled' for i in range(1000)]

df_trades = pd.DataFrame(dict2)
# df_trades
````

````python
from sqlalchemy import create_engine
engine = create_engine('sqlite://')
df_users.to_sql('users', con=engine)
df_trades.to_sql('trades', con=engine)
connection = engine.connect()
stmt = "SELECT u.City, COUNT(u.User_id) FROM users AS u JOIN trades AS t ON u.User_id = t.User_id WHERE DATE(t.Timestamp) = '2016-12-01' AND t.Status = 'cancelled' GROUP BY u.City ORDER BY 2 DESC LIMIT 5"
rs = connection.execute(stmt)
df = pd.DataFrame(rs.fetchall())
df
````

# Question 4: Modeling

Please use python, R or a similar tool for answering this question.
In order to improve user retention and lower churn, the growth team at Robinhood is interested in understanding why and which users withdraw money from their Robinhood account. A user is considered churned when their equity value (amount of money in Robinhood account) falls below $10 for a period of 28 consecutive calendar days or longer.
Using the datasets given below answer the following questions:

### a) What percentage of users have churned in the data provided? [4 points]

````python
import pandas as pd

df1 = pd.read_csv('equity_value_data.csv')

df1.info()
````

````python
df1.head()
````

````python
total_users = df1['user_id'].nunique()
print(total_users)
````

````python
df1[df1['close_equity'] < 10]
````

````python
print(df1['timestamp'].min())
print(df1['timestamp'].max())
````

````python
df1['date'] = pd.to_datetime(df1['timestamp'])
df1 = df1.sort_values(by=['user_id', 'date'])
df1.head()
````

````python
users = {}

for idx, row in df1.iterrows():
    uid = row['user_id']
    date = row['date']

    if uid in users:
        users[uid].append(date)
    else:
        users[uid] = [date]
````

````python
date_max = df1['date'].max()
twenty_eight_days = pd.Timedelta('28 days')

churn_user = []
counts = 0
for key, val in users.items():
    uid = key
    list_date = sorted(users[uid])
    for i in range(len(list_date) - 1):
        if list_date[i+1] - list_date[i] >= twenty_eight_days:
            churn_user.append(uid)
            counts += 1
            break
    if date_max - list_date[-1] >= twenty_eight_days:
        churn_user.append(uid)
        counts += 1
print(counts)
````

````python
print(counts / total_users * 100.)
````

### b) Build a classifier that given a user with their features assigns a churn probability for every user and predicts which users will churn. How well does your classifier perform? State any metrics you deem important here. Based on the classifier output classify each user in the dataset as churned or not churned. [5 points]

````python
selected_features = df3.columns.difference(['churn', 'user_id'])
print(selected_features)
````

````python
from sklearn.model_selection import train_test_split
X = df3[selected_features].values
y = df3['churn'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

````python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score

def get_performance_metrics(y_train, p_train_pred, y_test, p_test_pred, threshold=0.5):
    metric_names = ['AUC', 'Accuracy', 'Precision', 'Recall', 'f1-score']
    
    metric_values_train = [roc_auc_score(y_train, p_train_pred),
                           accuracy_score(y_train, p_train_pred > threshold),
                           precision_score(y_train, p_train_pred > threshold),
                           recall_score(y_train, p_train_pred > threshold),
                           f1_score(y_train, p_train_pred > threshold)]
    
    metric_values_test = [roc_auc_score(y_test, p_test_pred),
                          accuracy_score(y_test, p_test_pred > threshold),
                          precision_score(y_test, p_test_pred > threshold),
                          recall_score(y_test, p_test_pred > threshold),
                          f1_score(y_test, p_test_pred > threshold)]
    
    all_metrics = pd.DataFrame({'metrics': metric_names,
                                'train': metric_values_train,
                                'test': metric_values_test},
                                columns=['metrics', 'train', 'test'])

    all_metrics.set_index('metrics')
    print(all_metrics)
````

````python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def plot_roc_curve(y_train, p_train_pred, y_test, p_test_pred):
    roc_auc_train = roc_auc_score(y_train, p_train_pred)
    fpr_train, tpr_train, _ = roc_curve(y_train, p_train_pred)
    
    roc_auc_test = roc_auc_score(y_test, p_test_pred)
    fpr_test, tpr_test, _ = roc_curve(y_test, p_test_pred)
    
    lw=2
    plt.figure()
    plt.plot(fpr_train, tpr_train, color='green', linewidth=lw, label='ROC Train (AUC = %0.4f)' % roc_auc_train)
    plt.plot(fpr_test, tpr_test, color='darkorange', linewidth=lw, label='ROC Test (AUC = %0.4f)' % roc_auc_test)
    plt.plot([0, 1], [0, 1], color='navy', linewidth=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc='lower right')
    plt.show()
````

````python
def plot_feature_importance(model):
    df_feature_importance = pd.DataFrame()
    df_feature_importance['feature'] = selected_features
    df_feature_importance['importance'] = model.feature_importances_
    df_feature_importance.sort_values('importance', inplace=True)
    
    ax = df_feature_importance.plot(kind='barh', figsize=(20, 10))
    t = np.arange(len(df_feature_importance['feature']))
    ax.set_yticks(t)
    ax.set_yticklabels(df_feature_importance['feature'])
    plt.show()
````

````python
def train_test_model(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    
    y_train_pred = clf.predict(X_train)
    y_train_prob = clf.predict_proba(X_train)[:, 1]
    
    y_test_pred = clf.predict(X_test)
    y_test_prob = clf.predict_proba(X_test)[:, 1]
    
    get_performance_metrics(y_train, y_train_prob, y_test, y_test_prob)
    plot_roc_curve(y_train, y_train_prob, y_test, y_test_prob)
````

````python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.0, penalty='l2')
train_test_model(lr, X_train, y_train, X_test, y_test)
````

````python
coef_values = zip(selected_features, lr.coef_.flatten()) # connect feature names and values
df_coeffs = pd.DataFrame(list(coef_values))
df_coeffs.columns = ['Feature', 'Coefficient']
df_coeffs.sort_values(by='Coefficient', ascending=False, inplace=True)
df_coeffs
````

````python
from sklearn.ensemble import RandomForestClassifier
paras_rf = {'n_estimators': 50,
            'max_features': 'auto',
            'criterion': 'gini',
            'max_depth': 20,
            'min_samples_split': 2,
            'min_samples_leaf': 20,
            'random_state': 0,
            'n_jobs': -1}
rf = RandomForestClassifier(**paras_rf)

train_test_model(rf, X_train, y_train, X_test, y_test)
````
